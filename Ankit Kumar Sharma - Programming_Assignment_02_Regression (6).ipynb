```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import collections

# Generate meaningful data
np.random.seed(0)  # For reproducibility
x_values = np.linspace(60, 75, 25)  # Range of human height in inches
noise = np.random.normal(0, np.sqrt(20), 25)  # Gaussian noise
t_values = 3.86 * x_values - 110.42 + noise

# Plot scatter plot of data
plt.figure(figsize=(10, 6))
plt.scatter(x_values, t_values, color='blue', label='Noisy Data')
plt.title('Human Height vs. Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.legend()
plt.grid(True)
plt.show()

# Weight estimation through error surface
def Error(w, t, x):
    t_est = w[0] + w[1] * x
    error = np.sum((t - t_est) ** 2) / len(t)
    return error

# Plot 3D error surface and contour plots
w0_range = np.linspace(-120, -100, 100)
w1_range = np.linspace(3, 5, 100)
W0, W1 = np.meshgrid(w0_range, w1_range)
J = np.zeros(W0.shape)

for i in range(W0.shape[0]):
    for j in range(W0.shape[1]):
        J[i, j] = Error([W0[i, j], W1[i, j]], t_values, x_values)

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(121, projection='3d')
surf = ax.plot_surface(W0, W1, J, cmap='viridis')
plt.title('Error Surface')
plt.xlabel('w0')
plt.ylabel('w1')

# Contour plot
ax2 = fig.add_subplot(122)
contour = ax2.contour(W0, W1, J, 50)
plt.title('Error Contour')
plt.xlabel('w0')
plt.ylabel('w1')
plt.colorbar(surf, ax=ax, orientation='horizontal')
plt.show()

# Least squares approach to estimate the weights
def LinearRegression(x, t):
    X = np.vstack((np.ones_like(x), x)).T
    w_opt = np.linalg.inv(X.T @ X) @ X.T @ t
    return w_opt

w_opt = LinearRegression(x_values, t_values)
t_est = w_opt[0] + w_opt[1] * x_values

# Plot the estimated line on top of the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, t_values, color='blue', label='Noisy Data')
plt.plot(x_values, t_est, color='red', label='Estimated Line')
plt.title('Estimated Line with Least Squares')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (pounds)')
plt.legend()
plt.grid(True)
plt.show()

# The optimal weights estimated using the least squares approach are shown on the plot.
# These can be directly compared with the weights estimated through the error surface approach.
```
